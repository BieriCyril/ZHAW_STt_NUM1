# Visualisierung der Gauß-ähnlichen Funktion

In diesem Notebook visualisieren wir die Funktion $f(x) = e^{-x^2 / \sigma}$ für verschiedene Werte von $\sigma$.
Diese Funktion ist eng verwandt mit der Glockenkurve der Normalverteilung.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sorgt dafür, dass Plots direkt im Notebook angezeigt werden
%matplotlib inline
```

### Definition des Bereichs und der Parameter
Wir definieren den Definitionsbereich $x \in [-2, 2]$ und wählen drei verschiedene Werte für $\sigma$: $1/4$, $1/3$ und $1/2$.

```python
# 1. x-Bereich definieren (400 Punkte für eine glatte Kurve)
x = np.linspace(-2, 2, 400)

# 2. Sigma-Werte festlegen
sigmas = [1/4, 1/3, 1/2]
```

### Erstellung des Graphen
Nun iterieren wir über die $\sigma$-Werte, berechnen die Funktionswerte und plotten das Ergebnis.

```python
plt.figure(figsize=(10, 6))

for sigma in sigmas:
    # Berechnung von f(x) = e^(-x^2 / sigma)
    y = np.exp(-(x**2) / sigma)
    
    # Plot mit Label für die Legende
    plt.plot(x, y, label=f'$\sigma$ = {sigma:.2f}')

# Achsenbeschriftungen und Layout
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Graph der Funktion $f(x) = e^{-x^2 / \sigma}$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
