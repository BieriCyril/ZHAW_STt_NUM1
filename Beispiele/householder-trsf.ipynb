# Householder Spiegelung
Eine Householder-Spiegelung ist eine Geradenspiegelung

$$
    \vec{x} \mapsto H\, \vec{x}
$$

die *einen* bestimmten Vektor $\vec{a}$ auf die erste Koordinatenachse abbildet, also

$$
    H\, \vec{a} = |\vec{a}| \cdot \vec{e}_1 = \Big( |\vec{a}|, 0, \ldots,0 \Big)
$$

Da Geradenspiegelungen orthogonale (lineare) Abbildungen sind, ist die zugehörige Matrix $H$ eine orthogonale Matrix.

#### Konstruktion der Matrix H
Die Matrix $H$ spiegelt an der Winkelhalbierenden (Ebene) zwischen dem Vektor $\vec{a}$ und dem Einheitsvektor. Der Normalenvektor auf diese Ebene ist gegeben durch

$$
    \vec{n} = \vec{a} \pm |\vec{a}| \cdot \vec{e}_1
$$

Dies sind offenbar zwei Vektoren -- tatsächlich sind beide Varianten möglich. Die Matrix der Spiegelung ist dann gegeben durch

$$
    H = E - 2 \, \frac{\vec{n} \, \vec{n}^T}{|\vec{n}|^2}
$$

Dabei steht im Zähler das **äussere Produkt** von $\vec{n}$ mit sich selber:

$$
    \Big( \vec{n} \, \vec{n}^T \Big)_{ij} = n_i \, n_j
$$

und $E$ ist die Einheitsmatrix passender Grösse.

In dem interaktiven Plot kann man den Vektor $\vec{a}$ in der Ebene drehen. Ohne ipywidgets kann man aber auch die Funktion *householder* direkt aufrufen.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive
```

```python
# plot der Ergebnisse
def plotHouseholder(a, b, n):
    plt.plot([0, a[0]], [0, a[1]])
    plt.plot([-1.2*n[1], 1.2*n[1]], [1.2*n[0], - 1.2*n[0]], color = "orange", linestyle = "--") # Richtungsvektor
    plt.plot([0, b[0]], [0, b[1]], color = "red") 
    plt.plot([a[0], b[0]], [a[1], b[1]], color = [.6,.6,.6], linestyle = '--')
    plt.hlines(0, -1.5, 1.5, color = [.8,.8,.8], linestyle = '--')
    plt.axis("equal")
    plt.show()
    
# Berechnung der Householder-Transformation
# alpha legt den Vektor a = (cos alpha, sin alpha) fest
# sgn (1 oder -1) wählt die Variante für den Normalenvektor
def householder(a, sgn = -1):

    dim = a.shape[0] # das ist in diesem Beispiel immer = 2, aber...
    n = a.copy()  # a kopieren
    n[0] += sgn * np.linalg.norm(a) # Normalenvektor der Spiegelung
    n /= np.linalg.norm(n)    # normieren
    H = np.eye(dim) - 2 * np.outer(n, n) # Householder Transformation
    return H, n

def f(alpha, sgn = -1):
    # Einheitskreis zeichnen, nur für Skalierung
    p = np.linspace(0, 2*np.pi, 100)
    plt.plot(np.cos(p), np.sin(p), color = [.5,.5,1], linestyle = '--')
    
    # Vektor, der gespiegelt werden soll -> entspricht der jeweiligen Spalte der Matrix A
    a = np.array([np.cos(alpha), np.sin(alpha)]) 
    H, n = householder(a, sgn)   
    plotHouseholder(a, H@a, n)

```

```python
interactive_plot = interactive(f, alpha=(-np.pi, np.pi,1e-3), sgn = (-1, 1, 2) )
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
