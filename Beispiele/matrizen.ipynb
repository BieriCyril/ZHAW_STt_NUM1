# Matrizen in python (bzw. in numpy)


```python
import numpy as np
```

Vektoren und Matrizen sind **numpy arrays**. Im Gegensatz zu anderen Aufzählungen (listen) enthalten Sie nur einen einzigen (numerischen) Datentyp. Dafür unterstützen sie diverse Matrix- und Vektoroperationen

```python
b = np.array([3,-1,2,4,-5.])
print("Beispiel für Vektor", b)

A = np.random.randint(low = 2, high = 8, size = (5,5) ) # ganzzahlige Zufallsmatrix
print("\nZufällige 3 x 3 Matrix \n", A, "\n")

c = np.zeros_like(b) # Nullen der gleichen Form wie b. Funktioniert auch für Matrizen
print("c = ", c)

M = np.zeros( shape = (3, 4)) # alternativ: direkte Angabe der gewünschten Dimensionen
print("\nNullmatrix \n", M, "\n")

I = np.arange(start = 3, stop = 10) # "Zählvektor" oder Indexvektor. In vielem ähnlich wie das builtin range
print("range: ", I)
```

### Elementare Operationen mit Matrizen und Vektoren 
Für **Matrixprodukte** gibt es den Operator @. Der Stern "*" arbeitet elementweise. Daneben gibt es Matrix- und Vektornormen, Transposition etc.

```python
print("b = ", b, "\n\n")
print("A * b = ", A @ b)
print("Skalarprodukt von b mit sich selber = ", b @ b) # python unterscheidet nicht zwischen Zeilen- und Spaltenvektoren!
print("punktweises Produkt von b mit sich selber = ", b * b)
print("Transposition: A^T = \n", A.T)
print("Addition: b - 3 * b", b - 3 * b)
print("Addition einer Konstanten ", b - 7)
print("Summe aller Elemente von b: ", np.sum(b))
print("Produkt aller Elemente von b: ", np.prod(b))
print("Maximales Element von b: ", np.max(b))
print("Index, der zu dem Max gehört: ", np.argmax(b))
print("Vektornorm von b: ", np.linalg.norm(b))
```

### Indices
Alle Indices in python starten bei 0. Es kann auch auf ganze Bereiche zugegriffen werden. Obere Grenzen sind exklusive

```python
print("b = ", b) 
i = 2
print("Element ", i, " von b: ", b[i])
o = 3
print("Elemente bis ", o, " von b: ", b[:o])
u = 1
print("Elemente von ", u, " bis ", o, " von b: ", b[u:o])
print("Alle Elemente von ", u, " von b: ", b[u:])
```

**Negative Indices** zählen vom Ende her. b[-1] gibt das letzte Element von b

```python
i = -3
print("b[", i, "] = ", b[i])
```

**Analoges gilt für Matrizen**

```python
print("A = \n", A,"\n\n")
i, j = 3, 2
print("A[",i,",",j,"] = ", A[i, j], "\n")
print("Alle Elemente der Spalte ", j, " von A:", A[:,j], "\n")
print("Submatrix A[",i,":,",j,":] = \n", A[i:, j:], "\n")
print("Alle Elemente der Spalte ", i, " ab Zeile ", j, ": ", A[j:, i])
```

**Diagonale und für untere und obere Dreiecksmatrizen**

```python
print("A = \n", A, "\n")
print("Hauptdiagonale von A", np.diag(A))
i = -1
print("Untere Dreiecksmatrix von A\n", np.tril(A, i)) #analog triu für obere Dreiecksmatrizen
```

**Die Einheitsmatrix**

```python
print("Einheitsmatrix \n", np.eye(5))
```

Aus A die untere Dreiecksmatrix ohne Hauptdiagonale extrahieren und eine Einheitsmatrix addieren:

```python
np.tril(A,-1) + np.eye(5)
```

**Zeilen einer Matrix vertauschen**
Zeilen einer Matrix können auch in der Form
$$
    A[i]
$$
angesprochen werden, d.h. eine Matrix ist (wie in C) ein array von Vektoren, auch möglich ist z.B.
$$
    I = [1, 3, 2], \quad A[I]   
$$

oder direkt

$$
    A[[1, 3, 2]]
$$

```python
print("A = \n", A, "\n")
i, j = 0, 3
print("Zeilen ", i, " und ", j, " von A:\n ", A[[i, j]], "\n")
print("Zeilen ", i, " und ", j, " von A vertauschen")
A[[i, j]] = A[[j, i]]
print("A = \n", A, "\n")
```

Für Arrays werden bei Übergabe an Funktionen oder durch "=" nur die Referenzen übergeben bzw. gleichgesetzt.
$$
    B  = A
$$

bedeutet, dass B auf die Daten von A zugreift. Wenn man eine Kopie möchte, geht das mit

$$
    B = A.copy()
$$

```python
print("b = ", b) 
v = b.copy()                     # v = b oder v = b.copy() verhalten sich unterschiedlich
print("v = ", v)
v[0] = -v[0]
print("b = ", b, "  v = ", v)
```

Das entsprechende gilt für Übergabe an Funktionen (_call by reference_)

```python
def f(M):
    M *= 2

v = np.array([1,2,3])
print("v = ", v)
f(v)
print("v = ", v)
```

Wenn man die ursprüngliche Matrix behalten will, muss man beim Aufruf oder in der Funktion eine Kopie erstellen

```python
def f(M0):
    M = M0.copy() # Daten kopieren
    M *= 2

v = np.array([1,2,3])
print("v = ", v)
f(v)
print("v = ", v)
```

Logische Operatoren und logisches Indexing geht auch:

```python
u = 3 
print("A > ", u, "\n", A > u, "\n")
print("A > ", u, ":\n", 1 * (A > u), "\n")
print("alle Elemente von A, die > ", u, " sind:", A[A>3])
print("Alles <= ", u, " auf Null setzen: \n", A * (A > u)) # "*" wirkt elementweise!
```

### List comprehension
python bietet praktische Shortcuts zur Erstellung von Listen an. 

```python
# Gegendiagonale der Matrix A, d.h. alle Elemente der Form A[i, n-i]
print("A = \n", A, "\n")
n = A.shape[0]
d = [A[i, -(i+1)] for i in range(n)] # i = 0: [0,-1], i = 1: [1, -2], etc.
print("Gegendiagonale: ", d)
print("... in umgekehrter Reihenfolge: ", np.flip(d))
print("z. B. die Summe aller Elemente der Gegendiagonalen", np.sum([A[i, -(i+1)] for i in range(n)]))
```
